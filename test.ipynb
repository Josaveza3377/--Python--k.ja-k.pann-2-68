```python
midterm = input("กรุณากรอกคะแนนกลางภาค: ")
final = input("กรุณากรอกคะแนนปลายภาค: ")
classs = input("กรุณากรอกคะแนนเข้าเรียน : ") 
midterm =float(midterm)
final = float(final)
classs= float(classs)
print ("คะแนนรวม:" ,midterm + final + classs)
total_score = (midterm + final + classs)
if total_score > 79:
    print("ผลการเรียน: A")
elif total_score >= 70:
    print("ผลการเรียน: B")
elif total_score >= 60:
    print("ผลการเรียน: C")
elif total_score >= 50:
    print("ผลการเรียน: D")
elif total_score < 49:
    print("ผลการเรียน: แย่")
else:
    print("แจ้งเตือน: กรุณาป้อนข้อมูลเป็นตัวเลขจำนวนเต็มเท่านั้น")
print(type(midterm))
print(type(final))
print(type(classs))
```
